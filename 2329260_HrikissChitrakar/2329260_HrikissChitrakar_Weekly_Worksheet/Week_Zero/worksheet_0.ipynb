````python
def length_conversion(value, conversion_type):
    """
    Convert length between meters and feet.

    Parameters:
    value (float): The value to be converted.
    conversion_type (str): The type of conversion ('m_to_ft' or 'ft_to_m').

    Returns:
    float: Converted value.
    """
    if conversion_type == "m_to_ft":
        return value * 3.28084  # 1 meter = 3.28084 feet
    elif conversion_type == "ft_to_m":
        return value / 3.28084  # 1 foot = 0.3048 meters
    else:
        return None

def weight_conversion(value, conversion_type):
    """
    Convert weight between kilograms and pounds.

    Parameters:
    value (float): The value to be converted.
    conversion_type (str): The type of conversion ('kg_to_lbs' or 'lbs_to_kg').

    Returns:
    float: Converted value.
    """
    if conversion_type == "kg_to_lbs":
        return value * 2.20462  # 1 kg = 2.20462 lbs
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462  # 1 lb = 0.453592 kg
    else:
        return None

def volume_conversion(value, conversion_type):
    """
    Convert volume between liters and gallons.

    Parameters:
    value (float): The value to be converted.
    conversion_type (str): The type of conversion ('L_to_gal' or 'gal_to_L').

    Returns:
    float: Converted value.
    """
    if conversion_type == "L_to_gal":
        return value * 0.264172  # 1 liter = 0.264172 gallons
    elif conversion_type == "gal_to_L":
        return value / 0.264172  # 1 gallon = 3.78541 liters
    else:
        return None

def main():
    """
    Main function to prompt user input and perform conversions.
    """
    print("Welcome to Unit Converter!")
    print("Select the conversion type:")
    print("1. Length (meters <-> feet)")
    print("2. Weight (kilograms <-> pounds)")
    print("3. Volume (liters <-> gallons)")

    try:
        choice = int(input("Enter your choice (1/2/3): "))
        if choice not in [1, 2, 3]:
            print("Invalid choice! Please select a valid option.")
            return

        value = float(input("Enter the value to convert: "))

        if choice == 1:
            print("1. Convert meters to feet")
            print("2. Convert feet to meters")
            length_choice = int(input("Enter your choice (1/2): "))
            conversion_type = "m_to_ft" if length_choice == 1 else "ft_to_m"
            result = length_conversion(value, conversion_type)

        elif choice == 2:
            print("1. Convert kilograms to pounds")
            print("2. Convert pounds to kilograms")
            weight_choice = int(input("Enter your choice (1/2): "))
            conversion_type = "kg_to_lbs" if weight_choice == 1 else "lbs_to_kg"
            result = weight_conversion(value, conversion_type)

        elif choice == 3:
            print("1. Convert liters to gallons")
            print("2. Convert gallons to liters")
            volume_choice = int(input("Enter your choice (1/2): "))
            conversion_type = "L_to_gal" if volume_choice == 1 else "gal_to_L"
            result = volume_conversion(value, conversion_type)

        print(f"Converted Value: {result:.4f}")

    except ValueError:
        print("Invalid input! Please enter a numeric value.")

# Run the program
if __name__ == "__main__":
    main()
````

````python
def calculate_sum(numbers):
    """
    Calculate the sum of a list of numbers using a for loop.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Sum of numbers.
    """
    total = 0
    for num in numbers:
        total += num
    return total

def calculate_average(numbers):
    """
    Calculate the average of a list of numbers using a for loop.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Average value.
    """
    total = 0
    count = 0
    for num in numbers:
        total += num
        count += 1
    return total / count if count > 0 else 0

def calculate_maximum(numbers):
    """
    Find the maximum number in a list using a for loop.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Maximum number.
    """
    max_value = numbers[0]
    for num in numbers:
        if num > max_value:
            max_value = num
    return max_value

def calculate_minimum(numbers):
    """
    Find the minimum number in a list using a for loop.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: Minimum number.
    """
    min_value = numbers[0]
    for num in numbers:
        if num < min_value:
            min_value = num
    return min_value

def main():
    """
    Main function to prompt user input and perform calculations.
    """
    print("Math Operations on a List of Numbers")
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = int(input("Enter your choice (1/2/3/4): "))
        if choice not in [1, 2, 3, 4]:
            print("Invalid choice! Please select a valid option.")
            return

        numbers = input("Enter numbers separated by spaces: ").split()
        numbers = [float(num) for num in numbers]

        if not numbers:
            print("Error: No numbers provided.")
            return

        if choice == 1:
            result = calculate_sum(numbers)
            operation = "Sum"
        elif choice == 2:
            result = calculate_average(numbers)
            operation = "Average"
        elif choice == 3:
            result = calculate_maximum(numbers)
            operation = "Maximum"
        else:
            result = calculate_minimum(numbers)
            operation = "Minimum"

        print(f"{operation} of given numbers: {result:.4f}")

    except ValueError:
        print("Invalid input! Please enter numeric values only.")


if __name__ == "__main__":
    main()
````

# 4.3 Exercise on Nested List

````python
def extract_every_other(lst):
    """
    Extracts every other element from the list, starting from the first element.
    """
    new_list = []
    i = 0
    while i < len(lst):
        new_list.append(lst[i])
        i += 2
    return new_list

def get_sublist(lst, start, end):
    """
    Returns a sublist from the given list, from start to end (inclusive).
    """
    new_list = []
    i = start
    while i <= end:
        new_list.append(lst[i])
        i += 1
    return new_list

def reverse_list(lst):
    """
    Reverses the list without slicing.
    """
    new_list = []
    i = len(lst) - 1
    while i >= 0:
        new_list.append(lst[i])
        i -= 1
    return new_list

def remove_first_last(lst):
    """
    Removes the first and last elements of the list.
    """
    new_list = []
    i = 1
    while i < len(lst) - 1:
        new_list.append(lst[i])
        i += 1
    return new_list

def get_first_n(lst, n):
    """
    Extracts the first `n` elements from a list.
    """
    new_list = []
    i = 0
    while i < n and i < len(lst):
        new_list.append(lst[i])
        i += 1
    return new_list

def get_last_n(lst, n):
    """
    Extracts the last `n` elements from a list.
    """
    new_list = []
    i = len(lst) - n
    while i < len(lst):
        new_list.append(lst[i])
        i += 1
    return new_list

def reverse_skip(lst):
    """
    Extracts elements in reverse order, starting from the second-to-last element and skipping one element.
    """
    new_list = []
    i = len(lst) - 2
    while i >= 0:
        new_list.append(lst[i])
        i -= 2
    return new_list


def get_user_list():
    """
    Prompts the user to enter the number of elements and the list elements.
    """
    while True:
        try:
            n = int(input("Enter the number of elements in the list: "))
            if n <= 0:
                print("Please enter a positive integer.")
                continue
            break
        except ValueError:
            print("Invalid input! Please enter a valid integer.")

    user_list = []
    print(f"Enter {n} numbers one by one:")
    for i in range(n):
        while True:
            try:
                num = int(input(f"Enter element {i + 1}: "))
                user_list.append(num)
                break
            except ValueError:
                print("Invalid input! Please enter a number.")

    return user_list

if __name__ == "__main__":
    user_list = get_user_list()

    print("\nOperations on the list:")
    print("Every other element:", extract_every_other(user_list))

    while True:
        try:
            start = int(input("\nEnter start index for sublist: "))
            end = int(input("Enter end index for sublist: "))
            if 0 <= start <= end < len(user_list):
                break
            else:
                print("Invalid range! Start should be <= end and within list size.")
        except ValueError:
            print("Invalid input! Please enter numbers.")

    print("Sublist:", get_sublist(user_list, start, end))
    print("Reversed list:", reverse_list(user_list))
    print("Without first & last:", remove_first_last(user_list))

    while True:
        try:
            n = int(input("\nEnter number of elements to extract from the start: "))
            if n > 0:
                break
            else:
                print("Please enter a positive integer.")
        except ValueError:
            print("Invalid input! Please enter a valid integer.")

    print("First", n, "elements:", get_first_n(user_list, n))

    while True:
        try:
            n = int(input("\nEnter number of elements to extract from the end: "))
            if n > 0:
                break
            else:
                print("Please enter a positive integer.")
        except ValueError:
            print("Invalid input! Please enter a valid integer.")

    print("Last", n, "elements:", get_last_n(user_list, n))
    print("Reverse skipping:", reverse_skip(user_list))
````

`Problem 4: Matrix Operations`

````python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A⁻¹ = I
A_inv = np.linalg.inv(A)
identity_check = np.dot(A, A_inv)
print("A * A⁻¹ = Identity Matrix:\n", identity_check)

# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nMatrix AB:\n", AB)
print("\nMatrix BA:\n", BA)
if not np.array_equal(AB, BA):
    print("\nAB is NOT equal to BA")

# 3. Prove (AB)ᵀ = Bᵀ * Aᵀ
AB_transpose = np.transpose(AB)
BT_AT = np.dot(B.T, A.T)
print("\n(AB)ᵀ:\n", AB_transpose)
print("\nBᵀ * Aᵀ:\n", BT_AT)
if np.array_equal(AB_transpose, BT_AT):
    print("\n(AB)ᵀ is equal to Bᵀ * Aᵀ")

# 4. Solve Linear Equations using Inverse Methods
# System of equations:
#  2x - 3y + z = -1
#  x - y + 2z = -3
#  3x + y - z = 9

A_matrix = np.array([[2, -3, 1],
                     [1, -1, 2],
                     [3, 1, -1]])

B_matrix = np.array([-1, -3, 9])

A_inv = np.linalg.inv(A_matrix)
solution = np.dot(A_inv, B_matrix)
print("\nSolution for Linear Equations (x, y, z):\n", solution)
````
